```python
def generate_sql_create_statement(df, table_name):
    sql_data_types = {
        'int64': 'INT',
        'float64': 'FLOAT',
        'object': 'TEXT'
    }
    columns = df.columns
    column_types = df.dtypes
    
    column_sql_types = [f"{column} {sql_data_types[str(dtype)]}" for column, dtype in zip(columns, column_types)]
    create_statement = f"CREATE TABLE {table_name} (\n" + ",\n".join(column_sql_types) + "\n);"
    
    return create_statement

# Example usage:
table_name = "YourTableName"
create_statement = generate_sql_create_statement(dataframe, table_name)
print(create_statement)
```
